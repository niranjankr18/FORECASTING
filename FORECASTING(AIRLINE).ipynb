# FORECASTING MODELS BUILDING

we will forecast last two years 2001 and 2002 so the last records will be test data and first 70 records will be training data

```python
dt_train=dt[0:70]
dt_test=dt[70:]
```

```python
print("length of the train data:",len(dt_train))
print("length of the test data:",len(dt_test))
```

```python
dt_train.head()
```

```python
dt_test.head()
```

# METHODS FOR FORECASTING



# 1.SIMPLE MOVING AVERAGE

```python
dt_ma=dt.copy()
```

```python
train_len=70
```

```python
dt_ma['ma_forecast']=dt_ma['Passengers'].rolling(12).mean()
```

```python
dt_ma['ma_forecast'][train_len:] = dt_ma['ma_forecast'][train_len-1]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#RMSE AND MAPE
rmse_ma = mean_squared_error(dt_test['Passengers'],dt_ma['ma_forecast'][train_len:],squared=False).round(2)
rmse_ma
```

# 2:-SIMPLE EXPONENTIAL SMOOTHING

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
model = SimpleExpSmoothing(dt_train['Passengers']).fit(smoothing_level=0.13)
pre_model=model.predict(start= dt_test.index[0],end = dt_test.index[-1])
```

```python
rmse_ses = mean_squared_error(dt_test['Passengers'],pre_model,squared=False).round(2)
rmse_ses
```

# 3.HOLT'S METHOD

```python
from statsmodels.tsa.holtwinters import Holt
```

```python
model2 = Holt(dt_train["Passengers"]).fit(smoothing_level=0.8, smoothing_slope=0.23)
pred_holt = model2.predict(start = dt_test.index[0],end = dt_test.index[-1])
```

```python
rmse_holt = mean_squared_error(dt_test['Passengers'],pred_holt,squared=False).round(2)
rmse_holt
```

# 4. Holt's Exponential Smoothing

a. Additive method

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing 
```

```python
hes_add = ExponentialSmoothing(dt_train['Passengers'],seasonal="add",trend="add",seasonal_periods=12).fit()
pred_add = hes_add.predict(start = dt_test.index[0],end = dt_test.index[-1])
```

```python
rmse_hes = mean_squared_error(dt_test['Passengers'],pred_add,squared=False).round(2)
rmse_hes
```

b. Multiplicative method

```python
hes_mul = ExponentialSmoothing(dt_train['Passengers'],seasonal="mul",trend="add",seasonal_periods=12).fit() 
pred_mul= hes_mul.predict(start = dt_test.index[0],end = dt_test.index[-1])
```

```python
rmse_mul = mean_squared_error(dt_test['Passengers'],pred_mul,squared=False).round(2)
rmse_mul
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
arima = ARIMA(dt_train['Passengers'], order=(1,1,1)).fit()
pred_arima= arima.predict(start = dt_test.index[0],end = dt_test.index[-1])
```

```python
rmse_arima = mean_squared_error(dt_test['Passengers'],pred_arima,squared=False).round(2)
rmse_arima
```

# Choosing the best model

```python
models=['simple_moving_average','simple_expo_smoothing','Holt','Holt_additive','Holt_multiplicative','ARIMA']
rmse=[rmse_ma,rmse_ses,rmse_holt,rmse_hes,rmse_mul,rmse_arima]
d={'Models':models,'Rmse':rmse}
dt_model=pd.DataFrame(d)
dt_model
```

```python
dt_model.sort_values(by='Rmse')
```

here holt's exponential smoothing with multicative seasonality and additive trend has the best and lowest rmse.
